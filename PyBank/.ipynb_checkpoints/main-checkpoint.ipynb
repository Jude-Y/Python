```python
from pathlib import Path
import csv

csvpath = Path('../PyBank/budget_data.csv')
date = []
profitandloss = []
monthly_change = []
line_num = 0
diff_list = []
months = []

with open(csvpath, 'r') as csvfile:
    
    csvreader = csv.reader(csvfile)
    header = next(csvreader)
          
        
     
    prev = next(csvreader)
    for column in csvreader:
           
        diff = int(column[1]) - int(prev[1])
        diff_list.append(diff)
        prev = column
        dates = column[0]
        date.append(dates)  
        

with open (csvpath, 'r') as newcsv:
    csvnew = csv.reader(newcsv)
    head=next(newcsv)
    for pnl in csvnew:
          
        profitloss = int(pnl[1])
        profitandloss.append(profitloss) 
        
        total = str(pnl[0])
        months.append(total)
           
total_months = 0
net_total = 0
average_change = 0
greatest_increase_profits = 0
greatest_decrease_profits = 0
count_net_total = 0
count_average = 0
     
        
total_months = len(months)
print(f'Total Months: {total_months}')


for num in profitandloss:
    net_total += int(num)
    count_net_total += 86
print(f'Net Profit/(Losses): ${net_total}')
                

for avg in diff_list:
    average_change += avg
    count_average += 1
average = round(average_change / len(diff_list),2)
print(f'Average Change over the entire period: (${average})')


for x in diff_list:
           
      
    if greatest_decrease_profits == 0:
        greatest_decrease_profits = x
    elif x < greatest_decrease_profits:
        greatest_decrease_profits = x
    elif x > greatest_increase_profits:
        greatest_increase_profits = x
   
   


      
greatest_increase_month = diff_list.index(max(diff_list))
greatest_month = date[greatest_increase_month]


greatest_decrease_month = diff_list.index(min(diff_list))
least_month = date[greatest_decrease_month]  

print(f"Greatest Increase in Profits over the entire period was on {greatest_month} for the amount of ${greatest_increase_profits}")
print(f"Greatest Decrease in Profits over the entire period was on {least_month} for the amount of (${greatest_decrease_profits})")  
```
